```python
# Define the Volatility Breakout strategy

def VolBreakout(PD_Dataframe, percentage: int):
  # Convert int to Percentage
  multiplier = percentage / 100

  # Calculate the Top and Bottom of the range to determine where positions will be taken
  top_range_calc = PD_Dataframe['High'][-1] * multiplier
  bottom_range_calc = PD_Dataframe['Low'][-1] * multiplier
  top_of_range = PD_Dataframe['Open'] + top_range_calc
  bottom_of_range = PD_Dataframe['Open'] - bottom_range_calc

  # Initiate a Column that counts the number of times a cross happens on above the range and below the range
  PD_Dataframe['Top of Range'] = round(top_of_range, 2)
  PD_Dataframe['Above Range'] = 0
  PD_Dataframe['Bottom of Range'] = round(bottom_of_range, 2)
  PD_Dataframe['Below Range'] = 0

  # Find the columns where a cross happens and updates the column with a 1 if there is a cross
  PD_Dataframe['Above Range'] = np.where(PD_Dataframe['Close'] > round(top_of_range, 2), 1, 0)
  PD_Dataframe['Below Range'] = np.where(PD_Dataframe['Close'] < round(bottom_of_range, 2), 1, 0)

  return(PD_Dataframe)

```

```python
spy_bars = pd.read_csv(
    Path("/Users/andrewcrawford/Desktop/Algo-trading-project/data/SPY.csv"),
    index_col="Date",
    parse_dates=True,
    infer_datetime_format=True)

spy_plot = VolBreakout(spy_bars, 1)

vix_bars = pd.read_csv(
    Path("/Users/andrewcrawford/Desktop/Algo-trading-project/data/VIX.csv"),
    index_col="Date",
    parse_dates=True,
    infer_datetime_format=True)

vix_plot = VolBreakout(vix_bars, 1)

btc_bars = pd.read_csv(
    Path("/Users/andrewcrawford/Desktop/Algo-trading-project/data/BTC.csv"),
    index_col="Date",
    parse_dates=True,
    infer_datetime_format=True)

btc_plot = VolBreakout(btc_bars, 5)
```

```python
# Visualize exit position relative to close price
exit = btc_plot[btc_plot['Below Range'] == 1.0]['Close'].hvplot.scatter(
    color='green',
    marker='v',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize entry position relative to close price
entry = btc_plot[btc_plot['Above Range'] == 1.0]['Close'].hvplot.scatter(
    color='blue',
    marker='^',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

btc_close = btc_plot[['Close']].hvplot(
    line_color='red',
    ylabel='Price in $',
    width=1000,
    height=400
)

plot = exit * entry * btc_close

plot.opts(
    title="BOV Entry and Exit Points"
)
```
